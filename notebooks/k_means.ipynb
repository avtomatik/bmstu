```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
```

```python
np.random.seed(40)
```

```python
X, y = make_blobs(
    n_samples=200,
    n_features=2,
    centers=4,
    cluster_std=.7,
    shuffle=True
)
plt.title("Generated Points")
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.grid()
```

```python
X
```

```python
y
```

```python
k_means = KMeans(n_clusters=4, init='random', n_init=10, max_iter=100, tol=1e-04)
k_means_prediction = k_means.fit_predict(X)
```

```python
k_means_prediction
```

```python
compared = {(pre, post) for pre, post in zip(y, k_means_prediction)}
compared
```

```python
plt.title("Clusterization Results")

plt.scatter(X[:, 0], X[:, 1], c=k_means_prediction)
plt.scatter(*k_means.cluster_centers_.T, marker='*', s=200, color='r')
plt.grid()
```

```python
distortions = []
for _ in range(1, 7):
    k_means = KMeans(
        n_clusters=_,
        init='random',
        n_init=10,
        max_iter=300,
        tol=1e-04
    )
    k_means_prediction = k_means.fit_predict(X)
    plt.title(f"Clusterization Results with {_=} Clusters")
    plt.scatter(X[:, 0], X[:, 1], c=k_means_prediction)
    plt.scatter(*k_means.cluster_centers_.T, marker='*', s=200, color='r')
    plt.grid()
    plt.show()

    distortions.append(k_means.inertia_)
```

# Elbow Rule

```python
plt.plot(range(1, 7), distortions, marker='o')
plt.xlabel("Clusters' Count")
plt.xlabel("Distortion")
plt.grid()
plt.show()
```
