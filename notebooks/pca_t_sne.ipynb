```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets, decomposition, manifold
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
pca = decomposition.PCA()
pca.fit(X)
```

```python
explained_variance = pca.explained_variance_
explained_variance
```

```python
ev_sum = sum(explained_variance)
q = np.array([ev / ev_sum for ev in explained_variance])
```

```python
q
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_ratio
```

```python
pca.singular_values_
```

```python
covmat = np.cov(np.transpose(X))
covmat
```

```python
A = np.linalg.eig(covmat)
A
```

```python
eig_vectors = A[1]
eig_vectors
```

```python
main_components = np.transpose(eig_vectors[:, :2])
main_components
```

```python
X_new = np.dot(main_components, np.transpose(X))
np.transpose(X_new)
```

```python
from collections import OrderedDict
from functools import partial
from time import time

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets, manifold
```

```python
n_points = 2000
X, color = datasets.make_s_curve(n_points, random_state=10)
n_components = 2
```

```python
fig = plt.figure(figsize=(30, 15))
fig.suptitle(f"Manifold Learning (t-SNE) with {n_components} points", fontsize=14)
ax = fig.add_subplot(251, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
ax.view_init(4, -72)

methods = OrderedDict()
methods["t-SNE perplexity=5"] = TSNE(
    n_components=n_components,
    init='random',
    random_state=0,
    perplexity=5.
)
methods["t-SNE perplexity=15"] = TSNE(
    n_components=n_components,
    init='random',
    random_state=0,
    perplexity=15.
)
methods["t-SNE perplexity=30"] = TSNE(
    n_components=n_components,
    init='random',
    random_state=0,
    perplexity=30.
)
methods["t-SNE perplexity=50"] = TSNE(
    n_components=n_components,
    init='random',
    random_state=0,
    perplexity=50.
)

for _, (label, method) in enumerate(methods.items()):
    t0 = time()
    y = method.fit_transform(X)
    t1 = time()
    print(f"{label}: {t1 - t0:,.2f} sec")
    ax = fig.add_subplot(2, 5, 2 + _ + (_ > 3))
    ax.scatter(y[:, 0], y[:, 1], c=color, cmap=plt.cm.Spectral)
    ax.set_title(f"{label}: {t1 - t0:,.2f} sec")
    ax.axis("tight")

plt.show()
```
