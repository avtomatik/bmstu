```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
X = np.array([
    [0, 0],
    [1, 0],
    [0, 1],
    [1, 1],
    [2, 1],
    [1, 2],
    [2, 2],
    [3, 2],
    [6, 6],
    [7, 6],
    [8, 6],
    [6, 7],
    [7, 7],
    [8, 7],
    [9, 7],
    [7, 8],
    [8, 8],
    [9, 8],
    [8, 9],
    [9, 9],
])
```

```python
plt.scatter(X[:, 0], X[:, -1])
plt.grid()
```

```python
k_means = KMeans(n_clusters=2)
```

```python
?KMeans
```

```python
k_means.fit(X)
```

```python
k_means_clstr_cntrs = k_means.cluster_centers_
k_means_clstr_cntrs
```

```python
k_means.labels_
```

```python
plt.scatter(X[:, 0], X[:, -1], c=k_means.labels_, marker='^', cmap='rainbow')
plt.scatter(k_means_clstr_cntrs[:, 0], k_means_clstr_cntrs[:, -1], marker='*', cmap='rainbow')
plt.grid()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris
```

```python
X = iris.data
y = iris.target
```

```python
k_means_iris = KMeans(n_clusters=3).fit(X)
k_means_iris.cluster_centers_
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=k_means_iris.labels_, marker='o', cmap='rainbow')
plt.grid()
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y, marker='*', cmap='rainbow')
plt.grid()
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, 2], c=k_means_iris.labels_, marker='o', cmap='rainbow')
plt.grid()
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, 2], c=y, marker='*', cmap='rainbow')
plt.grid()
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, -1], c=k_means_iris.labels_, marker='o', cmap='rainbow')
plt.grid()
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, -1], c=y, marker='*', cmap='rainbow')
plt.grid()
```
